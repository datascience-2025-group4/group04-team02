```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSV oder TSV-Datei einlesen
# Beispiel: df = pd.read_csv("ab_ag.tsv", sep="\t")
df = pd.read_csv("../data/uniprot_data.tsv", sep="\t")

# Nur relevante Spalten auswählen und leere Werte filtern
df = df[["Organism", "Protein names"]].dropna()

# nur die 10 häufigsten werte
#top_species = df["antigen_species"].value_counts().nlargest(10).index
#df = df[df["antigen_species"].isin(top_species)]

# Kreuztabelle erstellen (Zählung jeder Kombination)
cross_tab = pd.crosstab(df["Organism"], df["Protein names"])

# Optional: relative Häufigkeit oder Normalisierung
# cross_tab = cross_tab.div(cross_tab.sum(axis=1), axis=0)

# Heatmap plotten
plt.figure(figsize=(12, 6))
sns.heatmap(cross_tab, cmap="viridis", annot=True, fmt="d", linewidths=0.5)
plt.title("Frequency of Organism × Protein names")
plt.xlabel("Antigen Species")
plt.ylabel("IGHV Subclass (Organism)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
