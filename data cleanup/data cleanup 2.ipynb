```python
import pandas as pd


# path to dataframe
df = pd.read_csv("../data/ab_ag.tsv", sep="\t")
df

# Arbeitskopie fr cleanup 
df_clean = df.copy()
df_clean
```

```python
print(df_clean["resolution"].dtype)
```

```python
# Ungltige Werte -> NaN
df_clean["resolution"] = pd.to_numeric(df_clean["resolution"], errors="coerce")  # Ungltige Werte  NaN
df_clean = df_clean[df_clean["resolution"] <= 3.2]                                     # Nur  3.2 behalten
df_clean = df_clean.dropna(subset=["resolution"])  

```

```python
# Drop rows with missing entries in following column
required_cols = ["Hchain", "Lchain", "antigen_chain", "antigen_species"]
df_clean = df_clean.dropna(subset=required_cols)

# Drop entries with missing or placeholder chains (like 'NA' as string)
df_clean = df_clean[~df_clean["Hchain"].isin(["NA"])]
df_clean = df_clean[~df_clean["Lchain"].isin(["NA"])]
df_clean = df_clean[~df_clean["antigen_chain"].isin(["NA"])]
df_clean = df_clean[~df_clean["antigen_species"].isin(["NA"])]


#to exclude single chain constructs
# df = df[(df["Hchain"] != "NA") & (df["Lchain"] != "NA")]

print(df_clean.columns.tolist())

# drop columns we don't need
df_clean = df_clean.drop(columns=["authors", "date", "antigen_het_name", "short_header", "compound", "model", "heavy_species", "light_species"])
df_clean = df_clean.drop(columns=["r_free", "r_factor", "engineered", "affinity", "delta_g", "affinity_method", "temperature", "pmid"])

df_clean
```
